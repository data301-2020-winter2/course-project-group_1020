```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
raw_data = pd.read_csv("path_to_the_file")

YearOfAttack = raw_data["iyear"]
Losses = raw_data["nkill"]

new_dataset = {
    "Year":YearOfAttack,
    "Losses":Losses
}

YearAndLosses = pd.DataFrame(new_dataset)
YearAndLosses["Losses"] = YearAndLosses["Losses"].dropna()
YearAndLosses["Losses"] = pd.to_numeric(YearAndLosses["Losses"], downcast = 'integer')

YearAndLosses 
```

```python
groupedByYear = YearAndLosses.groupby('Year')
groupedByYear
```

```python
for Year, Year_df in groupedByYear:
    print(Year)
    print(Year_df)
    
decade_attacks()
```

**Terrorist attacks by decades compared**

```python
# Calculate Amount of Terrorist Attacks each year.
def decade_attacks():
    seventies = 0
    eighties = 0
    nineties = 0
    zeroes = 0
    tenths = 0

    for i in range (0,10):
        seventies += (len(groupedByYear.get_group(1970+i)))

    for i in range (0,10):
        eighties += (len(groupedByYear.get_group(1980+i)))

    for i in range (0,10):
        if (1990 + i == 1993):
            nineties += 0
        else:
            nineties += (len(groupedByYear.get_group(1990+i)))
    
    for i in range (0,10):
        zeroes += (len(groupedByYear.get_group(2000+i)))
    
    for i in range (0,8):
        tenths += (len(groupedByYear.get_group(2010+i)))
    

plot_decade_attacks()
 
```

```python
#Creating a pie chart
def plot_decade_attacks():
    plt.figure(figsize=(8,8))

    years = ["1970s","1980s","1990s","2000s","2010s"]
    attacks = [seventies, eighties, nineties, zeroes, tenths]

    plt.pie(attacks, labels=years, autopct = '%0.1f%%')
    plt.title("Terrorist Attacks By Decades")
    plt.show()
```

```python
#def losses_each_decade():
    
data = {'Year':[],
       'Losses':[]}

losses_df = pd.DataFrame(data)
    
for Year, Year_df in groupedByYear:
    count = 0
    for losses in Year_df["Losses"]:
        if (pd.isna(losses)):
            count += 0
        else:
            count += losses
        
    new_row = {'Year':Year, 'Losses':count}
    #print(new_row)
    losses_df = losses_df.append(new_row, ignore_index = True)

losses_df
```

```python
#Plotting histogram

plt.figure(figsize=(10,5))

#plt.scatter(losses_df['Year'],losses_df['Losses'])
plt.bar(losses_df['Year'], losses_df['Losses'], width = 0.85, align = 'center')

plt.title("Losses caused by terrorist attacks 1970 - 2017")
plt.xlabel('Years')
plt.ylabel('Losses')

plt.show()
```
