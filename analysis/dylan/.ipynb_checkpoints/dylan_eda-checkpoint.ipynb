<h3>Data Collection</h3>
The <i>Global Terrorism Database</i> (GTD) is maintained by the National Consortium for the <i>Study of Terrorism and Responses to
Terrorism</i> (START) at the University of Maryland.<br>
The GTD is the product of several phases of data collection efforts, each relying on
publicly available, unclassified source materials including: media articles, electronic
news archives, and some existing data sets, books, journals, and legal documents.<br>
The data collection efforts first started with <i>Pinkerton Global Intelligence Service</i> (PGIS) from 1970 to 1997. The <i>Center
for Terrorism and Intelligence Studies</i> (CETIS) continued data collection from 1998 to 2008 and expanded the scope of the data then the <i>Institute for the Study of Violent Groups</i> (ISVG) picked it up until 2011. All ongoing data collection is being continued by START themselves.

```python
countplt, ax = plt.pyplot.subplots(figsize = (9,5))
ax = (sns.countplot(y=df['region'],
                    order=df['region'].value_counts().index,
                    palette='rocket')
         .set_title('Number of Terrorist Attacks')
     )
plt.pyplot.gca().spines['top'].set_visible(False)
plt.pyplot.gca().spines['right'].set_visible(False)
plt.pyplot.ylabel('Region')
```

'Middle East & North Africa' and 'South Asia' contain the largest sum of terrorist attacks.<br>
Combined, these two regions constitute half the terrorist attacks between 1970 to 2017. 

```python
plt.pyplot.figure(figsize=(22,5))
c_t = df[['region','year']]
country_terrorism = (c_t.groupby(c_t.columns.tolist())
                        .size().reset_index()
                        .rename(columns={0:'records'})
                    )
country_terrorism = pd.pivot_table(country_terrorism, index='region', columns='year', values='records')
plt.pyplot.title('Number of Terroristic Attacks')
sns.heatmap(country_terrorism, robust=True, xticklabels=3)
```

The regions that see a large number of terrorist attacks between 1970 to 1998 are most likely due to these countries establishing their governance and rule.<br>
These countries saw a lot of insurrections, political instability, and new countries forming or splitting.<br>
After 2001 with the terrorist attacks in America, we can see the war on terrorism start and escalate.

```python
t_n = df[['region', 'country', 'nationality']]
t_n = t_n.assign(records = "international")
t_n.loc[t_n['country'] == t_n['nationality'], "records"] = "domestic"

countplt, ax = plt.pyplot.subplots(figsize = (12,7))
ax = (sns.countplot(y=t_n['region'],
                    hue=t_n['records'],
                    order=t_n['region'].value_counts().index,
                    palette='icefire',
                    dodge=True
                   )
         .set_title('Terrorism Against Targets of Domestic or Non-domestic Nationality')
     )
plt.pyplot.gca().spines['top'].set_visible(False)
plt.pyplot.gca().spines['right'].set_visible(False)
plt.pyplot.legend(loc="center right", fontsize='medium', title='Nationality', title_fontsize='medium')
```

The data we have contains the country a terrorist attack took place, and the nationality of the target of that terrorist attack.<br>
The ratio's of terrorist attacks against domestic or non-domestic targets is about 10:1 or greater for each region except Western Europe.<br>
Western Europe has an almost 1:1 ratio of terrorist attacks agaisnt domestic and non-doemstic targets.<br>
In the next section we explore this oddity deeper.

<h3>The Oddity of Western Europe</h3>

```python
t_n = df[['year', 'region', 'country', 'nationality']]
t_n = t_n[t_n['country'] != t_n['nationality']]

plt.pyplot.figure(figsize=(22,5))
c_t = t_n[['region','year']]
country_terrorism = (c_t.groupby(c_t.columns.tolist())
                        .size().reset_index()
                        .rename(columns={0:'records'})
                    )
country_terrorism = pd.pivot_table(country_terrorism, index='region', columns='year', values='records')
plt.pyplot.title('Terroristic Attacks of Non-domestic Nationality')
sns.heatmap(country_terrorism, robust=True, xticklabels=3)
```

When we look at the number of terrorist attacks by non-domestic targets only, Western Europe has a far larger number of attacks than any other regions.<br>
Whats peruliar about Eu is that the number of attacks are almost stagnant over the years until 2001.<br> 
The Middle East & North Africa have a sizeable number but may just be due to the large number of total attacks they have.<br>
After 2001, there is once again a big dip in terrorist activity but starts rising again over a decade later.

```python
t_n = df[['year', 'region', 'country', 'nationality', 'targettype', 'attacktype', 'weapontype', 'weaponsubtype']]
t_n = t_n[t_n['region'] == "Western Europe"]
t_n = t_n[t_n['country'] != t_n['nationality']]

plt.pyplot.figure(figsize=(22,8))
c_t = t_n[['targettype','year']]
country_terrorism = (c_t.groupby(c_t.columns.tolist())
                        .size().reset_index()
                        .rename(columns={0:'records'})
                    )
country_terrorism = pd.pivot_table(country_terrorism, index='targettype', columns='year', values='records')
plt.pyplot.title('Type of Targets in Western Europe of Non-domestic Nationality')
sns.heatmap(country_terrorism, robust=True, xticklabels=3)
```

When we look at the types of non-domestic targets, we would expect tourists and airports/aircrafts to be the largest targets but they aren't.<br>
This plot is actually very typical, with a majority of targets being the general population of citizens, businesses, and government enforcement/employees.<br>
What we can find out though is that a majority of current terrorist attacks are on private citizens and property.

```python
plt.pyplot.figure(figsize=(22,4))
f_n = df[df['region'] == "Western Europe"]
c_t = f_n[['attacktype','year']]
country_terrorism = (c_t.groupby(c_t.columns.tolist())
                        .size().reset_index()
                        .rename(columns={0:'records'})
                    )
country_terrorism = pd.pivot_table(country_terrorism, index='attacktype', columns='year', values='records')
plt.pyplot.title('Attack Types in Western Europe')
sns.heatmap(country_terrorism, robust=True, xticklabels=3)
```

Bombing/Explosions are still the majority attack type in Western Europe but assassinations take the second spot.<br>
Usually armed assaults are the second highest of terrorist attacks in regions but this may be due to the stric gun laws in Western Europe.<br>
Assassinations were huge before the 2000s then they became almost non existent. This may be due to the lack of political instability.

```python
countplt, ax = plt.pyplot.subplots(figsize = (9,5))
ax = (sns.countplot(y=df['attacktype'],
                    order=df['attacktype'].value_counts().index,
                    palette='rocket')
         .set_title('Attack Types of Terroristic Attacks')
     )
plt.pyplot.gca().spines['top'].set_visible(False)
plt.pyplot.gca().spines['right'].set_visible(False)
plt.pyplot.ylabel('Attack Type')
```

For attack types around the world, Bombing/Explosion types are half of all attacks, Armed Assault is a quarter of attacks, and Assassination is an eighth of attacks.<br>
<br>
In conclusion, Western Europe probably has half of its attacks as non-domestic because of the European Union. The European Union allows citizens in the countries apart of it to quite easily find work and housing in any of the countries. Immigration is a lot easier compared to other parts of the world which may mean there is a mix of nationalities across Western Europe. The 1:1 ratio of domestic and non-domestic targets may simply be due to the random distribution of mixed nationalities that makes up Western Europe.

<h3>Ransoms and Kidnappings</h3>

```python
ax = (sns.barplot(y=df['region'], 
                  x=df['ransom']*100
                 )
         .set_title('Percentage of Ransoms')
     )
plt.pyplot.gca().spines['top'].set_visible(False)
plt.pyplot.gca().spines['right'].set_visible(False)
plt.pyplot.xlabel('ransom %')
```

Not a lot of terrorist attacks involve ransoms, on average they only occur 1.5% of the time. 

```python
ax = (sns.barplot(y=df['region'], 
                  x=df['ishostkid']*100)
         .set_title('Percentage of Kidnappings')
     )
plt.pyplot.gca().spines['top'].set_visible(False)
plt.pyplot.gca().spines['right'].set_visible(False)
plt.pyplot.xlabel('kidnapping %')
```

Kidnappings are a lot more frequent compared to ransoms, occurring just under 8% of the time.<br>
From the 1.5% ransoms, we can gather that 20% of kidnappings are for money while the rest of kidnappings are for other nefarious purposes.

```python
k_n = df[['region', 'country', 'ishostkid', 'kidnaptocountry']]
k_n = k_n[k_n['ishostkid'] == 1]
k_n = k_n.assign(records = "out of borders")
k_n.loc[k_n['country'] == k_n['kidnaptocountry'], "records"] = "stayed in borders"

countplt, ax = plt.pyplot.subplots(figsize = (12,7))
ax = (sns.countplot(y=k_n['region'],
                    hue=k_n['records'],
                    order=k_n['region'].value_counts().index,
                    palette='icefire',
                    dodge=True
                   )
         .set_title('Count of Kidnappings that Stayed in or out of Borders')
     )
plt.pyplot.gca().spines['top'].set_visible(False)
plt.pyplot.gca().spines['right'].set_visible(False)
plt.pyplot.legend(loc="center right", fontsize='medium', title='Kidnapped', title_fontsize='medium')
```

```python
countplt, ax = plt.pyplot.subplots(figsize = (9,5))
ax = (sns.countplot(y=k_n['records'],
                    order=k_n['records'].value_counts().index,
                    palette='rocket')
         .set_title('Attack Types of Terroristic Attacks')
     )
plt.pyplot.gca().spines['top'].set_visible(False)
plt.pyplot.gca().spines['right'].set_visible(False)
plt.pyplot.ylabel('Kidnapped')
```

Here we can see how many kidnappings stayed within the country and how many continued out of the country.<br>
The regions with a 1:1 ratio of kidnappings in or out of the border imply that half of the kidnappings are for the purpose of needing hostages or ransom money while the other half are probably for trafficking.<br>
Regions such as Asia and Africa have a far majority of their kidnappings leaving the country, most likely for trafficking purposes.<br>
We can reinforce this reasoning by comparing the 20% of kidnappings occurring within borders to the 20% of kidnappings with ransoms.
